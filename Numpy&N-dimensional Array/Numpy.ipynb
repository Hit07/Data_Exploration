# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

```python
np.matmul(a1,b1)
```

```python
print(a1 @ b1)
```

# Manipulating Images as ndarrays


```python
import pooch
from scipy.datasets import face


img = face()
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
img.shape
```

```python
img.ndim
```

Convert the image to black and white. The values in our `img` range from 0 to 255. 


```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
b_and_w = (img/255.0) @ grey_vals
```

```python
plt.imshow(b_and_w,cmap='gray')
plt.axis('off')
```

Manipulate the images by doing some operations on the underlying ndarrays? See if you can change the values in the ndarray so that:



```python
plt.imshow(np.flip(b_and_w),cmap='gray')
plt.axis('off')
```

```python
plt.imshow(np.flip(img))
plt.axis('off')
```

```python
plt.imshow(np.rot90(img))
plt.axis('off')
```

```python
solarized = 255-img
plt.imshow(solarized)
plt.axis('off')
```
